```python
from bs4 import BeautifulSoup
import requests
import re

import time
```

```python
boyNAmes = []
girlNAmes = []
for i in range(ord('A'),ord('Z')+1):

    pageNum = 1
    print('Char :',chr(i),'.'*15)
    print('Boy Names')
    
    while(True):
        basePage = 'http://www.bachpan.com/Bengali-Boy-Names-'+chr(i)+'.aspx?page='+str(pageNum)
        response = requests.get(basePage)
        soup = BeautifulSoup(response.content, "html5lib")
        names = [link.get_text() for link in soup.find_all('td',{'class':['c1']})]
        if not len(names) :
            break
        boyNAmes.extend(names)
        pageNum +=1
        time.sleep(1)
        
        
    basePage = 'https://www.momjunction.com/baby-names/islam/boy/starting-with-'+chr(i)
    while(basePage):
        response = requests.get(basePage)
        soup = BeautifulSoup(response.content, "html5lib")
        names = [link.get_text() for link in soup.find_all('tbody')[0].find_all('a')[::2]]
        boyNAmes.extend(names)
        try:
            basePage = soup.find('ul',class_='pagination col-sm-8 babynames_pagination').find_all('li')[-1].find('a').get('href')
        except:
            break
        time.sleep(1)
        

    basePage = 'https://www.urdupoint.com/islamic-names/alpha-sort/boys/'+chr(i).lower()+'/1'
    while(True):
        response = requests.get(basePage)
        soup = BeautifulSoup(response.content, "html5lib")
        names = [link.get_text() for link in soup.find_all('ul',{'class':['list_item_wrap']})[0].find_all('a')]
        boyNAmes.extend(names)
        try :
            tmp = soup.find('a', title = 'Next Page').get('href')
        except:
            break
        if tmp == basePage:
            break
        else :
            basePage = tmp
    
    
    basePage = 'https://www.searchtruth.com/baby_names/names.php?ntype=m&find=2&letter='+chr(i)
    response = requests.get(basePage)
    soup = BeautifulSoup(response.content, "html5lib")
    try :
        names =  [link.get_text() for link in soup.find_all('tbody')[4].find_all('a')]
        boyNAmes.extend(names)
    except:
        pass

    time.sleep(1)
     
        
    basePage = 'http://www.abohomanbangla.com/muslim/boy/'+chr(i+32)+'.html'
    response = requests.get(basePage)
    soup = BeautifulSoup(response.content, "html5lib")
    try :
        names =  [(tr.find_all('td')[1].get_text()).replace(' ', '') for tr in soup.find_all('tbody')[5].find_all('tr')]
        boyNAmes.extend(names)
    except:
        pass
    
    time.sleep(1)
    
    print(len(boyNAmes))

    
    
    
    print('Girl Names')
    
    pageNum = 1
    while(True):

        basePage = 'http://www.bachpan.com/Bengali-Girl-Names-'+chr(i)+'.aspx?page='+str(pageNum)
        response = requests.get(basePage)
        soup = BeautifulSoup(response.content, "html5lib")
        names = [link.get_text() for link in soup.find_all('td',{'class':['c1']})]
        if not len(names) :
            break
        girlNAmes.extend(names)
        pageNum +=1
        time.sleep(1)
        
    
    basePage = 'https://www.momjunction.com/baby-names/islam/girl/starting-with-'+chr(i)
    while(basePage):
        response = requests.get(basePage)
        soup = BeautifulSoup(response.content, "html5lib")
        names =[link.get_text() for link in soup.find_all('tbody')[0].find_all('a')[::2]]
        girlNAmes.extend(names)
        try:
            basePage = soup.find('ul',class_='pagination col-sm-8 babynames_pagination').find_all('li')[-1].find('a').get('href')
        except:
            break
        time.sleep(1)
        
        
    basePage = 'https://www.urdupoint.com/islamic-names/alpha-sort/girls/'+chr(i).lower()+'/1'
    while(True):
        response = requests.get(basePage)
        soup = BeautifulSoup(response.content, "html5lib")
        names = [link.get_text() for link in soup.find_all('ul',{'class':['list_item_wrap']})[0].find_all('a')]
        girlNAmes.extend(names)
        try :
            tmp = soup.find('a', title = 'Next Page').get('href')
        except:
            break
            
        if tmp == basePage:
            break
        else :
            basePage = tmp
        
        time.sleep(1)
        
    basePage = 'https://www.searchtruth.com/baby_names/names.php?ntype=f&find=2&letter='+chr(i)
    response = requests.get(basePage)
    soup = BeautifulSoup(response.content, "html5lib")
    try :
        names =  [link.get_text() for link in soup.find_all('tbody')[4].find_all('a')]
        girlNAmes.extend(names)
    except:
        pass

    time.sleep(1)
        

    basePage = 'http://www.abohomanbangla.com/muslim/girl/'+chr(i+32)+'.html'
    response = requests.get(basePage)
    soup = BeautifulSoup(response.content, "html5lib")
    names =  [(tr.find_all('td')[1].get_text()).replace(' ', '') for tr in soup.find_all('tbody')[5].find_all('tr')]


    girlNAmes.extend(names)
    print(len(girlNAmes))

    time.sleep(1)
```

```python
banglaNames ={'male':boyNAmes,'female':girlNAmes}
```

```python
from json import dump
```

```python
with open('banglaNames.json','w') as outfile:
    dump(banglaNames, outfile)
    outfile.close()
```
